```python
import os
import requests
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import chromedriver_autoinstaller

class InstagramBot:
    def __init__(self, email, password):
        chromedriver_autoinstaller.install()
        self.browser = webdriver.Chrome()
        self.email = email
        self.password = password

    def signIn(self):
        self.browser.get('https://www.instagram.com/accounts/login/')
        time.sleep(2)

        emailInput = self.browser.find_element(By.NAME, 'username')
        passwordInput = self.browser.find_element(By.NAME, 'password')

        emailInput.send_keys(self.email)
        passwordInput.send_keys(self.password)
        passwordInput.send_keys(Keys.ENTER)
        time.sleep(5)

        try:
            wait = WebDriverWait(self.browser, 10)
            wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "sqdOP.yWX7d.y3zKF"))).click()
            time.sleep(2)
            wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "aOOlW.HoLwm"))).click()
        except Exception as e:
            print(f"Exception occurred: {e}")

    def scrape_data_and_download_images(self):
        profiles = []
        datetimelist = []
        SCROLL_PAUSE_TIME = 2

        last_height = self.browser.execute_script("return document.body.scrollHeight")
        for _ in range(5):
            try:
                name_elements = WebDriverWait(self.browser, 10).until(
                    EC.presence_of_all_elements_located((By.XPATH, "//a[contains(@href, '/')]"))
                )
                date_elements = WebDriverWait(self.browser, 10).until(
                    EC.presence_of_all_elements_located((By.XPATH, "//time[@class='_1o9PC Nzb55']"))
                )

                for n in name_elements:
                    profile = n.get_attribute("href").split('/')[-2]
                    profiles.append(profile)
                
                for d in date_elements:
                    date = d.get_attribute("datetime")
                    datetimelist.append(date)

                self.browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(SCROLL_PAUSE_TIME)

                new_height = self.browser.execute_script("return document.body.scrollHeight")
                if new_height == last_height:
                    break
                last_height = new_height
            except Exception as e:
                print(f"Exception occurred: {e}")
                break

        print(f"Profiles: {profiles}")
        for profile in profiles:
            self.download_images(profile)

        self.save_data(profiles, datetimelist)

    def download_images(self, profile):
        self.browser.get(f'https://www.instagram.com/{profile}/')
        time.sleep(3)

        images = self.browser.find_elements(By.TAG_NAME, 'img')
        print(f"Found {len(images)} images for profile {profile}")

        save_dir = '/Users/pratyushkumargulzari/Desktop/insta_data'
        if not os.path.exists(save_dir):
            os.makedirs(save_dir)

        for idx, image in enumerate(images):
            src = image.get_attribute('src')
            print(f"Downloading image {idx} from {src}")
            response = requests.get(src)
            if response.status_code == 200:
                with open(f'{save_dir}/{profile}_{idx}.jpg', 'wb') as file:
                    file.write(response.content)
                print(f"Saved image {idx} for profile {profile}")

    def save_data(self, profiles, datetimelist):
        save_dir = '/Users/pratyushkumargulzari/Desktop/insta_data'

        if not os.path.exists(save_dir):
            os.makedirs(save_dir)

        try:
            new_profile = [profile.replace("See All", "").strip() for profile in profiles if profile.strip()]
            df1 = pd.DataFrame(new_profile, columns=['profile_name'])
            df1.to_csv(os.path.join(save_dir, 'profiles.csv'), index=False)

            new_date = [date.split('T')[0] for date in datetimelist if date]
            df2 = pd.DataFrame(new_date, columns=['date'])
            df2.to_csv(os.path.join(save_dir, 'datetime.csv'), index=False)

            print("Data saved successfully.")
        except Exception as e:
            print(f"Exception occurred while saving data: {e}")

if __name__ == "__main__":
    bot = InstagramBot('kashmere.gate.shenanigans', 'kashmeregate06')
    bot.signIn()
    time.sleep(5)
    bot.scrape_data_and_download_images()
```
